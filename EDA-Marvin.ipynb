## Data Preprocessing

```python
#import sys
#!{sys.executable} -m pip install pandas
import pandas as pd
data = pd.read_csv('/Users/clairebentzen/Desktop/MDAS/ADS 502 - Applied Data Mining/archive/2015_data.csv')

# identify shape of data
print('Number of columns = %d' % (data.shape[0])) # number of columns
print('Number of attributes = %d' % (data.shape[1])) # number of rows
data.head()
```

```python
# subset desired columns
columns = ['resident_status', 'month_of_death', 'sex', 'detail_age', 'place_of_death_and_decedents_status', 'marital_status', 'day_of_week_of_death', 'manner_of_death', 'method_of_disposition', 'autopsy', 'race', ]
data = data[columns]
data.head()
```

```python
# missing values
print('Number of missing values:')
for col in data.columns: 
    print('\t%s: %d' % (col,data[col].isna().sum())) 
```

```python
# import sys
# !{sys.executable} -m pip install matplotlib
import matplotlib
%matplotlib inline

data['detail_age'] = pd.to_numeric(data['detail_age']) # convert column to numeric
data.boxplot(figsize=(20,3)) # create boxplot on each column
```

```python
# duplicate data
dups = data.duplicated() # returns duplicated rows
print('Number of duplicate rows = %d' % (dups.sum()))
```

```python
data = data.drop_duplicates() # drop duplicate rows
print('Number of rows after discarding duplicates = %d' % (data.shape[0]))
```

```python
# sample 25% of the data
sample = data.sample(frac=0.25, random_state=1)
print('Number of rows after random sampling = %d' % (data.shape[0]))
```

```python
from pandas.api.types import is_numeric_dtype

num_col = ['detail_age']
# calculate summary stats for numeric columns
for col in num_col:
    print('%s:' % (col))
    print('\t Mean = %.2f' % data[col].mean())
    print('\t Standard deviation = %.2f' % data[col].std())
    print('\t Minimum = %.2f' % data[col].min())
    print('\t Maximum = %.2f' % data[col].max())
```

```python
# counts frequency of each attribute class
sample['resident_status'].value_counts()
sample['month_of_death'].value_counts()
sample['place_of_death_and_decedents_status'].value_counts()
sample['sex'].value_counts()
sample['marital_status'].value_counts()
sample['method_of_disposition'].value_counts()
sample['race'].value_counts()
sample['manner_of_death'].value_counts()
```
